#22
Derive a new variable, saturated_ fat_per_gram, by dividing the amount of saturated fat by the weight in grams.<br>
a.  Sort the data set by saturated_ fat_per_gram and produce a listing of the five food items highest in saturated fat per gram.<br>
b.  Which food has the most saturated fat per gram?<br>

```python
nutrition['saturated_fat_per_gram'] = nutrition['saturated_fat'] / nutrition['weight_in_grams']

nutrition_sortBy_fatPerGram = nutrition.sort_values(['saturated_fat_per_gram'], ascending=False)
nutrition_sortBy_fatPerGram.head()
```

#24
Standardize the field saturated_ fat_per_gram. <br>
Produce a listing of all the food items that are outliers at the high end of the scale.<br>
How many food items are outliers at the low end of the scale?

```python
from scipy import stats
nutrition['saturated_fat_per_gram_z'] = stats.zscore(nutrition['saturated_fat_per_gram'])

nutrition_fat_z_sort = nutrition.sort_values(['saturated_fat_per_gram_z'], ascending=False)
nutrition_fat_z_sort.head(n=5)
```

```python
nutrition_outliers_fat = nutrition_fat_z_sort.query('saturated_fat_per_gram_z > 3 | saturated_fat_per_gram_z < -3')
np.shape(nutrition_outliers_fat)
```

```python
nutrition_outliers_fat
```

```python
print('15 foods are outliers at the high end of the scale.' +
      '\nNo foods is outlier at the low end of the scale.')
```

#27
Determine whether any outliers exist for the education field.

```python
adult['edu_z'] = stats.zscore(adult['education']) # Standardize

adult_edu_z_sort = adult.sort_values(['edu_z'], ascending=False) # Sort edu_z

edu_outliers = adult_edu_z_sort.query('edu_z > 3 | edu_z < -3') # Find outliers
edu_outliers
```

#28
Do the following for the age field.<br>
a.  Standardize the variable.<br>
b.  Identify how many outliers there are and identify the most extreme outlier.

```python
adult['age_z'] = stats.zscore(adult['age'])
adult_age_z_sort = adult.sort_values(['age_z'], ascending=False)

age_outliers = adult_age_z_sort.query('age_z > 3 | age_z < -3')
age_outliers.sort_values(['age_z'], ascending=False).head()
```

```python
sum([(i>3 or i<-3) for i in adult["age_z"]])
```

```python
print('There are %.0f outliers in age field.' % (age_outliers.shape[0]))
```
