```python
#a
job_list = list(set(bank_train["job"]))

crosstab_30 = pd.crosstab(bank_train['response'], bank_train['job'])
crosstab_30_norm = round(crosstab_30.div(crosstab_30.sum(0), axis = 1)*100, 1)

dict_job = {}
bank_train['job2'] = bank_train['job']
for i in job_list: 
    if crosstab_30_norm[i]['yes'] < 10:
        dict_job[i] = '0<10'
    elif crosstab_30_norm[i]['yes'] >= 10 and crosstab_30_norm[i]['yes'] <25:
        dict_job[i] = '10<25'
    elif crosstab_30_norm[i]['yes'] >= 25 and crosstab_30_norm[i]['yes'] <33:
        dict_job[i] = '25<33'
    else:
        pass

bank_train.replace({'job2' : dict_job}, inplace=True)
bank_train['job2']
```

```python
adult['age_binned'] = pd.cut(x= adult['age'], bins=[0, 30, 60, max(adult.age)], labels=["<30","30-60","over 60"], right=False)

crosstab_57 = pd.crosstab(adult['age_binned'], adult['income'])
crosstab_57norm = round(crosstab_57.div(crosstab_57.sum(1), axis=0)*100, 1)
crosstab_57norm.plot(kind='bar', xlabel = 'Age', ylabel = 'Proportion', title='Normalized Bar Graph of Age with Income overlay', stacked = True, figsize = (10,5))
```
